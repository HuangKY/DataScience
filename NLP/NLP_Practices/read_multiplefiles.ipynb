    * read multiple files in some specific folder

```python
import os
import glob

import pandas as pd
import numpy as np
```

```python
# Read files
# Ref: https://stackoverflow.com/questions/18262293/how-to-open-every-file-in-a-folder
# Ref: https://stackoverflow.com/questions/541390/extracting-extension-from-filename-in-python

df = pd.DataFrame()
for filefullpath in glob.glob('./data/points-separate*.csv'):
    
    # filefullpath = './data/points_separate_1.csv', ...
    fullfilename = os.path.basename(filefullpath)
    
    # fullfilename = 'points-separate_1.csv', ...
    filename, file_extension = os.path.splitext(fullfilename)
    
    # filename = 'points_separate_1', ...
    filename, sn = filename.split(sep='_')[0], filename.split(sep='_')[1]
    
    
    # read and concat the files
    df_read = pd.read_csv(filefullpath, sep=',')
    df_read['data_source'] = filename
    df_read['serial_number'] = sn
    df = pd.concat([df, df_read], axis=0)
    
df
```
