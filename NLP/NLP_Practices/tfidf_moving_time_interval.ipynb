# Introduction
In this notebook, I will try to separate a file to differenct files with some moving-time-interval approach.

```python
import pandas as pd
import numpy as np
import matplotlib as plt
from datetime import date, datetime, timedelta
from dateutil.relativedelta import relativedelta
import math
```

# Data Preprocessing

```python
# Read Data

df = pd.read_csv('./data/points.csv')
```

```python
df.head()
```

```python
# Change to datetime / date objects

df['datetime'] = df['datetime'].map(lambda x: datetime.strptime(x, "%Y/%m/%d %H:%M:%S"))
df['date'] = df['datetime'].map(lambda x: x.date())
```

# month_range_generator

```python
# create function month_range_generator
month_window_size = 3
# INPUT: month_window_size (INT)
# OUTPUT: all the segments (moving months) based on the month_window_size (2-D lists)

# Get min/max
min_date, max_date = df['date'].min(), df['date'].max()
print(min_date, max_date)
# Date_Trunc
min_date = min_date.replace(day=1)
max_date = max_date.replace(day=1) + relativedelta(months=1) # the first date of next month
print(min_date, max_date)

# Get data_diff and range
r = relativedelta(min_date, max_date)
month_ranges = abs(r.years * 12 + r.months)
num_segments = math.ceil(month_ranges / month_window_size)  # integer

# Create segments (no intersect)
lower_date = min_date
segments = []

for i in range(num_segments):
    
    upper_date = lower_date + relativedelta(months=month_window_size)
    
    if upper_date >= max_date:  # The final segment
        segments.append((i + 1, lower_date, max_date))
        break
        
    segments.append((i + 1, lower_date, upper_date)) # list of serialnum, lower, upper
    lower_date = upper_date
    
    
# Create segments (with intersection)
lower_date = min_date
upper_date = min_date # initialization
i = 0
overlap_segments = []
while upper_date < max_date:
    upper_date = lower_date + relativedelta(months=month_window_size)
    overlap_segments.append((i + 1, lower_date, upper_date)) # list of serialnum, lower, upper
    lower_date = lower_date + relativedelta(months=+1)
    i += 1
```

# Separate Files

```python
# Separate files  (no intersect)
for segment in segments:
    df_part = df[(df['date'] >= segment[1]) & (df['date'] < segment[2])]
    file_name = './output/corpus_' + str(segment[0]) + '_' + segment[1].strftime("%Y%m%d") + '_' + segment[2].strftime("%Y%m%d") + '.tsv'
    df_part.to_csv(file_name, header=True, index=False, sep='\t')
    print('file size is {}'.format(len(df_part)))
```

```python
# Separate files (with intersection)
for overlap_segment in overlap_segments:
    df_part = df[(df['date'] >= overlap_segment[1]) & (df['date'] < overlap_segment[2])]
    file_name = './output/corpus_overlap_' + str(overlap_segment[0]) + '_' + overlap_segment[1].strftime("%Y%m%d") + '_' + overlap_segment[2].strftime("%Y%m%d") + '.tsv'
    df_part.to_csv(file_name, header=True, index=False, sep='\t')
    print('file size is {}'.format(len(df_part)))
```
